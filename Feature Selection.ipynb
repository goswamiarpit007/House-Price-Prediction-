```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
pd.pandas.set_option('display.max_columns', None)
```

```python
mainDF=pd.read_csv(r'C:\Users\goswa\Documents\House price Prediction\AfterFeatureEngineering.csv')
```

```python
mainDF.head()
```

```python
X=mainDF.iloc[:,2:]

Y=mainDF['SalePrice']
```

```python
X.head()
```

```python
Y.head()
```

```python
# Feature Selection
feature_select=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_select.fit(X,Y)
```

```python
z=feature_select.get_support()
z
```

```python
selected_feat = X.columns[(feature_select.get_support())]
len(selected_feat)
```
