### Handling Missing Values

```python
#Missing Categorical Features
missingCategorical=[feature for feature in mainDF.columns if mainDF[feature].isnull().sum()>=1 and mainDF[feature].dtype=='O']
for feature in missingCategorical:
    print(feature,'--------',np.round(100*(mainDF[feature].isnull().mean()),3),'%')
```

```python
#Handling categorical missing values
def replace_cat_feature(dataset,features_nan):
    data=mainDF.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

mainDF=replace_cat_feature(mainDF,missingCategorical)

mainDF[missingCategorical].isnull().sum()
```

```python
mainDF.head()
```

```python
#Finding missing values in numerical features
numericalMissing=[feature for feature in mainDF.columns if mainDF[feature].isnull().sum()>=1 and mainDF[feature].dtype!='O']
for feature in numericalMissing:
    print(feature,'------',np.round(100*(mainDF[feature].isnull().mean()),3),'%')
```

```python
#replacing missing values with median values since there are outliers in these dependent features
for feature in numericalMissing:
    medianValue=mainDF[feature].median()
    mainDF[feature+'nan']=np.where(mainDF[feature].isnull(),1,0)
    mainDF[feature].fillna(medianValue,inplace=True)
mainDF[numericalMissing].isnull().sum()    
```

```python
mainDF.head(20)
```

```python
# Handling Year features related in EDA 
yearFeature=['YearBuilt','YearRemodAdd','GarageYrBlt']
for feature in yearFeature:
    mainDF[feature]-mainDF['YrSold']-mainDF[feature]
mainDF.head()    
```

```python
numericalFeatures=[feature for feature in mainDF.columns if mainDF[feature].dtype!='O']
discrete=[feature for feature in mainDF[numericalFeatures].columns if len(mainDF[feature].unique())<25]
continuous=[feature for feature in mainDF[numericalFeatures].columns if feature not in discrete+yearFeature+['Id']]

for feature in continuous:
    if feature=='SalePrice':
        pass
    else:
        mainDF[feature]=np.log1p(mainDF[feature])
```

```python
mainDF.head()
```

### Handling Rare Categorical Feature

removing categorical variables less than 1 % of observation

```python
categoricalFeatures=[feature for feature in mainDF.columns if mainDF[feature].dtype=='O']
categoricalFeatures
```

```python
for feature in categoricalFeatures:
    temp=mainDF.groupby(feature)['SalePrice'].count()/len(mainDF)
    temp_df=temp[temp>0.01].index
    mainDF[feature]=np.where(mainDF[feature].isin(temp_df),mainDF[feature],'Rare_var')
```

```python
mainDF.head()
```

## Feature Scaling

```python
# Converting Categorical Feature to Numerical Value
from sklearn.preprocessing import LabelEncoder
for feature in categoricalFeatures:
    le = LabelEncoder()   
    mainDF[feature ]= le.fit_transform(mainDF[feature ])
```

```python
mainDF.head()
```

```python
#Feature Scaling (NORMALIZATION)
featureToBeScaled=[feature for feature in mainDF.columns if feature not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(mainDF[featureToBeScaled])
```

```python
scaler.transform(mainDF[featureToBeScaled])
```

```python
mainDF = pd.concat([mainDF[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(mainDF[featureToBeScaled]), columns=featureToBeScaled)],
                    axis=1)
```
